````
#HESAP MAKINESI
def carpma(sayi1, sayi2):
    sonuc = 0.0
    for _ in range(int(sayi1)):
        sonuc += sayi2
    return sonuc

def karekok(x):
    tahmin = 0.0
    artim = 0.01
    while True:
        kare = carpma(tahmin, tahmin)
        if kare > x:
            break
        tahmin += artim
    return round(tahmin, 2)

def us_al(taban, us):
    tam_kisim = int(us)
    ondalik_kisim = us - tam_kisim

    # Tam kƒ±smƒ± hesapla
    sonuc = 1.0
    for _ in range(tam_kisim):
        sonuc = carpma(sonuc, taban)

    # Eƒüer ondalƒ±k kƒ±sƒ±m varsa (≈üimdilik sadece 0.5 destekliyoruz)
    if round(ondalik_kisim, 2) == 0.5:
        kok = karekok(taban)
        sonuc = carpma(sonuc, kok)

    return round(sonuc, 2)

# √ñrnekler
print(us_al(4, 1.5))  # 4^1.5 = 4 * sqrt(4) = 4 * 2 = 8
print(us_al(9, 0.5))  # sqrt(9) = 3
print(us_al(2, 3.5))  # 2^3 * sqrt(2) = 8 * ~1.41 = ~11.3

#LIST COMPREHENSION
print([i*i for i in range(0,10,2)])
print([i / 10 if i * 2 > 10 else 0 for i in range(0, 10, 2)])
````

###üß† Yeni Konu: Basit G√∂r√ºnt√º Filtresi (Bulanƒ±kla≈ütƒ±rma / Blur)
NumPy kullanarak 2D g√∂rsel √ºzerine ortalama filtresi uygulayacaƒüƒ±z. Bu, √ßevresindeki piksellerin ortalamasƒ±nƒ± alarak g√∂rseli yumu≈üatƒ±r (bulanƒ±kla≈ütƒ±rƒ±r).



````
# once fonksiyonu tanƒ±mla:
def blur(image):
  image = image.astype(np.float32)
  blurred_image=np.copy(image)
  height,width=image.shape
  for i in range(1,height-1):
    for j in range(1,width-1):
      blurred_image[i,j]=np.mean(image[i-1:i+2,j-1:j+2])
  return blurred_image

#filtreyi uygula ve gorsellestir.
blur_gorsel=blur(normalize_gorsel)
plt.imshow(blur_gorsel,cmap="gray")
plt.title("Blur G√∂r√ºnt√º")
plt.colorbar()
plt.show()
````

####üîÑ ƒ∞stersen bunlarƒ± da yapabilirsin (bonus)
üìå Kenar tespiti i√ßin np.abs(image[i+1, j] - image[i, j]) gibi farklƒ± filtreler dene.

üìà Filtre √∂ncesi ve sonrasƒ± ortalama, std gibi metrikleri kar≈üƒ±la≈ütƒ±r.

kenar tespiti (edge detection) mantƒ±ƒüƒ±nƒ± anlatacaƒüƒ±m ve np.abs(image[i+1, j] - image[i, j]) satƒ±rƒ±nƒ± doƒüru ≈üekilde bir filtre fonksiyonu i√ßine yerle≈ütireceƒüiz.

üéØ Ama√ß:
Pikselin kendisi ile altƒ±ndaki piksel arasƒ±ndaki farkƒ± alarak kenarlarƒ± ortaya √ßƒ±karmak.

######‚úÖ 1. Kenar Tespiti i√ßin Basit Fark Filtresi
Bu filtre, pikseller arasƒ±ndaki ani deƒüi≈üimleri tespit eder ‚Äî bu da genellikle bir kenar anlamƒ±na gelir.

````
def kose_tanimlama(image):
    image = image.astype(np.float32)
    kenar_tanimlama = np.zeros_like(image)
    height, width = image.shape

    for i in range(1, height - 1):
        for j in range(1, width - 1):
            dikey_fark = np.abs(image[i+1, j] - image[i, j])
            yatay_fark = np.abs(image[i, j+1] - image[i, j])
            kenar_tanimlama[i, j] = max(dikey_fark, yatay_fark)

    return kenar_tanimlama

# Fonksiyonu uygula
kenarlar = kose_tanimlama(normalize_gorsel)

# G√∂rselle≈ütir
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1) #1satƒ±r : 3 s√ºtun 1.grafik
plt.imshow(normalize_gorsel, cmap='gray')
plt.title("Orijinal G√∂r√ºnt√º")

plt.subplot(1, 3, 2) #1 satƒ±r :3 s√ºtun 2.grafik
plt.imshow(blur_gorsel, cmap='gray')
plt.title("Bulanƒ±k G√∂r√ºnt√º")

plt.subplot(1, 3, 3) #1 satƒ±r :3 s√ºtun 3. grafik
plt.imshow(kenarlar, cmap='gray')
plt.title("Kenar Tespit")

plt.tight_layout() #√ºst √ºste binme d√ºzensizlik varsa d√ºzeltir
plt.show()

# ƒ∞statistik
print("Bulanƒ±k G√∂r√ºnt√º Ortalamasƒ±:", blur_gorsel.mean())
print("Bulanƒ±k G√∂r√ºnt√º Std Sapmasƒ±:", blur_gorsel.std())

"""Eƒüer kod √ßalƒ±≈ümadƒ±ysa, muhtemelen ≈üu sorunlardan biri olu≈ümu≈ütur:

G√∂rselin (image) tipi int olduƒüu i√ßin, ortalamalarƒ± alƒ±rken tam sayƒ± b√∂lme hatasƒ± olabilir.

blurred[i, j] = np.mean(...) ifadesi float d√∂nd√ºr√ºr, ama blurred hala int dizisi olduƒüu i√ßin tip uyu≈ümazlƒ±ƒüƒ± olabilir.

Kod biraz yava≈ü √ßalƒ±≈üabilir √ß√ºnk√º i√ß i√ße d√∂ng√º var."""
````

######üß† NEDƒ∞R: E≈üikleme (Thresholding)?
E≈üikleme, piksel deƒüerlerinin belirli bir sƒ±nƒ±rƒ±n altƒ±nda veya √ºst√ºnde olup olmadƒ±ƒüƒ±na g√∂re i≈ülem yapƒ±lmasƒ±dƒ±r.

√ñrneƒüin:

Eƒüer kenar kuvveti 0.1‚Äôden b√ºy√ºkse: "bu bir kenardƒ±r"

Aksi halde: "kenar deƒüil"



````
def kenar_esikleme(edge_image, threshold=0.1):
    binary_koseler = np.zeros_like(edge_image)
    binary_koseler[edge_image > threshold] = 1.0
    return binary_koseler

kenar_gri = kose_tanimlama(normalize_gorsel)  # √ñnce kenarlarƒ± bul
kenar_belirgin = kenar_esikleme(kenar_gri, threshold=0.08)  # E≈üik uygula

plt.imshow(kenar_belirgin, cmap='gray')
plt.title("E≈üiklenmi≈ü Kenarlar")
plt.show()
````

````
# BONUS : Kontur ile KENAR √áƒ∞Zƒ∞Mƒ∞

plt.contour(kenar_gri, levels=[0.1], colors='red')
plt.title("Kontur ile Kenar √áizimi")
plt.show()
````

````
def kontrast_arttir(gorsel):
    return (gorsel - gorsel.min()) / (gorsel.max() - gorsel.min())
    return kontrast_arttir

plt.imshow(kontrast_arttir(normalize_gorsel),cmap="gray")
plt.title("Kontrastlƒ± G√∂r√ºnt√º")
plt.colorbar()
plt.show()
````

######üîç Sobel Filtresi ile Geli≈ümi≈ü Kenar Tespiti
üìå Nedir Bu?
Sobel filtresi, hem yatay hem de dikey kenarlarƒ± aynƒ± anda daha net tespit etmek i√ßin kullanƒ±lƒ±r.
Basit fark alma y√∂ntemlerinden daha g√º√ßl√ºd√ºr ve g√∂r√ºnt√ºdeki ani deƒüi≈üimleri daha iyi algƒ±lar.

üß† Temel Mantƒ±k:
Sobel, iki y√∂nl√º filtre uygular:


````
import matplotlib.pyplot as plt
from scipy.ndimage import convolve

# Sobel filtre √ßekirdekleri
sobel_x = np.array([[ -1, 0, 1],
                    [ -2, 0, 2],
                    [ -1, 0, 1]])

sobel_y = np.array([[ -1, -2, -1],
                    [  0,  0,  0],
                    [  1,  2,  1]])
#Bu matrisler, g√∂r√ºnt√ºye konvol√ºsyon i≈ülemi (kaydƒ±rƒ±larak √ßarpma) uygulanarak kenar bilgisi √ßƒ±karƒ±lƒ±r.

def sobel_kenar(image):
    image = image.astype(np.float32)
    gx = convolve(image, sobel_x)
    gy = convolve(image, sobel_y)
    sobel_magnitude = np.sqrt(gx**2 + gy**2)
    return sobel_magnitude / sobel_magnitude.max()  # normalize et

sobel_sonuc = sobel_kenar(normalize_gorsel)

plt.imshow(sobel_sonuc, cmap='gray')
plt.title("Sobel Kenar Tespiti")
plt.axis("off")
plt.show()
````

* Kavram	A√ßƒ±klama
* convolve	== G√∂r√ºnt√ºye filtreyi uygular (kaydƒ±rmalƒ± √ßarpma)
* sobel_x, sobel_y	==G√∂r√ºnt√ºde y√∂nl√º deƒüi≈üimi bulur
* sqrt(gx¬≤ + gy¬≤)	==Ger√ßek kenar b√ºy√ºkl√ºƒü√º
* normalize ==	G√∂rseli 0-1 arasƒ± parlaklƒ±ƒüa getirir

######üîç Canny Kenar Tespiti Nedir?
* Canny algoritmasƒ±, g√∂r√ºnt√ºdeki en doƒüru ve keskin kenarlarƒ± bulmak i√ßin geli≈ütirilmi≈ü bir y√∂ntemdir. Sobel'den daha akƒ±llƒ±dƒ±r √ß√ºnk√º:

üß† Adƒ±mlarƒ±

Yani:

````
#1 QAUSS FILTRESI UYGULAMA
from scipy.ndimage import gaussian_filter

blurred = gaussian_filter(normalize_gorsel, sigma=1.0)

plt.imshow(blurred, cmap='gray')
plt.title(" GAUSS FILTRESI ILE Bulanƒ±kla≈ütƒ±rƒ±lmƒ±≈ü G√∂r√ºnt√º ")
plt.colorbar()
plt.show()
````

````
#2 SOBEL ILE KENAR GUCU HESABI
from scipy.ndimage import convolve

# Sobel filtre √ßekirdekleri
sobel_x = np.array([[ -1, 0, 1],
                    [ -2, 0, 2],
                    [ -1, 0, 1]])

sobel_y = np.array([[ -1, -2, -1],
                    [  0,  0,  0],
                    [  1,  2,  1]])
#Bu matrisler, g√∂r√ºnt√ºye konvol√ºsyon i≈ülemi (kaydƒ±rƒ±larak √ßarpma) uygulanarak kenar bilgisi √ßƒ±karƒ±lƒ±r.

def sobel_kenar(image):
    image = image.astype(np.float32)
    gx = convolve(image, sobel_x)
    gy = convolve(image, sobel_y)
    sobel_magnitude = np.sqrt(gx**2 + gy**2)
    return sobel_magnitude / sobel_magnitude.max()  # normalize et

sobel_sonuc = sobel_kenar(blurred)

plt.imshow(sobel_sonuc, cmap='gray')
plt.title("Sobel Kenar Tespiti")
plt.axis("off")
plt.show()
````

````
#3 THRESHOLDING ile zayif-kuvvetli kenar ayrilir.
def kenar_esikleme(edge_image, threshold=0.1):
    binary_koseler = np.zeros_like(edge_image)
    binary_koseler[edge_image > threshold] = 1.0
    return binary_koseler

kenar_gri = kose_tanimlama(sobel_sonuc)  # √ñnce kenarlarƒ± bul
kenar_belirgin = kenar_esikleme(kenar_gri, threshold=0.08)  # E≈üik uygula

plt.imshow(kenar_belirgin, cmap='gray')
plt.title("E≈üiklenmi≈ü Kenarlar")
plt.show()
````

````
#Non-maximum suppression (incele≈ütirme) yapƒ±lƒ±r
"""NMS, kenarlarƒ± inceltmek i√ßin kullanƒ±lƒ±r.
Ama√ß: Kenar boyunca sadece en g√º√ßl√º pikseli (maksimum gradyan) bƒ±rakmak, diƒüerlerini sƒ±fƒ±rlamak.
Sobel veya diƒüer gradyan filtreleri kenar boyunca geni≈ü alanlar verir ‚Äî kalƒ±n √ßizgiler.
Ama biz yalnƒ±zca tam kenar hattƒ±nƒ± istiyoruz.
Bu y√ºzden, kenar boyunca sadece en y√ºksek deƒüeri koruruz, diƒüerlerini sƒ±fƒ±rlarƒ±z."""

def non_maximum_suppression(magnitude, gx, gy):
    height, width = magnitude.shape
    suppressed = np.zeros_like(magnitude)

    angle = np.arctan2(gy, gx) * 180.0 / np.pi
    angle[angle < 0] += 180  # A√ßƒ±yƒ± 0-180 aralƒ±ƒüƒ±na getir

    for i in range(1, height - 1):
        for j in range(1, width - 1):
            q = 255
            r = 255

            # 0 derece
            if (0 <= angle[i, j] < 22.5) or (157.5 <= angle[i, j] <= 180):
                q = magnitude[i, j + 1]
                r = magnitude[i, j - 1]
            # 45 derece
            elif (22.5 <= angle[i, j] < 67.5):
                q = magnitude[i + 1, j - 1]
                r = magnitude[i - 1, j + 1]
            # 90 derece
            elif (67.5 <= angle[i, j] < 112.5):
                q = magnitude[i + 1, j]
                r = magnitude[i - 1, j]
            # 135 derece
            elif (112.5 <= angle[i, j] < 157.5):
                q = magnitude[i - 1, j - 1]
                r = magnitude[i + 1, j + 1]

            if (magnitude[i, j] >= q) and (magnitude[i, j] >= r):
                suppressed[i, j] = magnitude[i, j]
            else:
                suppressed[i, j] = 0

    return suppressed

# Sobel y√∂nlerini de alacak ≈üekilde fonksiyonu g√ºncelle
def sobel_kenar_yonlu(image):
    image = image.astype(np.float32)
    gx = convolve(image, sobel_x)
    gy = convolve(image, sobel_y)
    magnitude = np.sqrt(gx**2 + gy**2)
    return magnitude / magnitude.max(), gx, gy

# Kenar g√ºc√º ve y√∂nlerini al
sobel_magnitude, gx, gy = sobel_kenar_yonlu(blurred)

# NMS uygula
nms_sonuc = non_maximum_suppression(sobel_magnitude, gx, gy)

# G√∂r√ºnt√ºle
plt.imshow(nms_sonuc, cmap='gray')
plt.title("Non-Maximum Suppression Sonucu")
plt.axis("off")
plt.show()
````

````
#Edge tracking by hysteresis (kenarlarƒ± birbirine baƒülama) uygulanƒ±r
"""Nedir Bu "Hysteresis"?
Bu adƒ±mda, elimizdeki kenarlarƒ± g√º√ßl√º ve zayƒ±f olarak sƒ±nƒ±flandƒ±rƒ±yoruz:

G√º√ßl√º kenar: Y√ºksek deƒüere sahip ‚Äî bunlar kesin kenardƒ±r.

Zayƒ±f kenar: Daha d√º≈ü√ºk deƒüere sahip ‚Äî ama eƒüer g√º√ßl√º kenara baƒülƒ±ysa, onu da tut."""

def hysteresis_thresholding(img, low_thresh=0.05, high_thresh=0.15):
    strong = 1.0
    weak = 0.5

    result = np.zeros_like(img)

    strong_i, strong_j = np.where(img >= high_thresh)
    weak_i, weak_j = np.where((img >= low_thresh) & (img < high_thresh))

    result[strong_i, strong_j] = strong
    result[weak_i, weak_j] = weak

    # Kom≈üulukla zayƒ±f kenarlarƒ± g√º√ßl√ºye baƒülƒ±ysa tut
    for i in range(1, img.shape[0]-1):
        for j in range(1, img.shape[1]-1):
            if result[i, j] == weak:
                # 8 kom≈üusuna bak
                if ((result[i+1, j-1] == strong) or (result[i+1, j] == strong) or (result[i+1, j+1] == strong)
                    or (result[i, j-1] == strong) or (result[i, j+1] == strong)
                    or (result[i-1, j-1] == strong) or (result[i-1, j] == strong) or (result[i-1, j+1] == strong)):
                    result[i, j] = strong
                else:
                    result[i, j] = 0
    return result

hysteresis_sonuc = hysteresis_thresholding(nms_sonuc, low_thresh=0.05, high_thresh=0.15)

plt.imshow(hysteresis_sonuc, cmap='gray')
plt.title("Hysteresis Thresholding Sonucu (Canny'nin Finali- ama el ile kodlanmƒ±≈ü)")
plt.axis("off")
plt.show()
````

######üß† Kƒ±saca Kontur Nedir?

G√∂rseldeki t√ºm ≈üekillerin √ßevresini d√∂ner dola≈üƒ±r gibi tanƒ±mlar.

````
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Kontur bul
contours, _ = cv2.findContours(canny_kenarlar, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Konturlarƒ± orijinal g√∂r√ºnt√º √ºzerine √ßizmek i√ßin bir kopya
kontur_gorsel = cv2.cvtColor(normalize_gorsel_uint8, cv2.COLOR_GRAY2BGR)
cv2.drawContours(kontur_gorsel, contours, -1, (0, 255, 0), 1)

for i, cnt in enumerate(contours[:10]):  # ƒ∞lk 10 kontur
    alan = cv2.contourArea(cnt)
    cevre = cv2.arcLength(cnt, True)
    print(f"Kontur {i+1}: Alan={alan:.2f}, √áevre={cevre:.2f}")

# G√∂rselle≈ütir
plt.figure(figsize=(10,5))
plt.imshow(kontur_gorsel)
plt.title(f"{len(contours)} adet kontur bulundu")
plt.axis("off")
plt.show()
````

######‚úÖ 1. Kontur Alanƒ±, √áevresi ve Merkezi Hesaplama, Kontur √ºst√ºnde Dikd√∂rtgen veya Daire √áizimi
***üß† Ne √ñƒürendik?***
* **cv2.contourArea:** Alan hesaplar.

* **cv2.arcLength:** √áevre hesaplar.

* **cv2.boundingRect:** En k√º√ß√ºk dikd√∂rtgen.

* **cv2.minEnclosingCircle:** En k√º√ß√ºk √ßember.

* **cv2.moments:** Nesnenin merkezini verir.

````
"""1. Kontur Alanƒ±, √áevresi ve Merkezi Hesaplama"""

for i, kontur in enumerate(contours):
    alan = cv2.contourArea(kontur)
    cevre = cv2.arcLength(kontur, True)

    # M (moment) hesapla ‚Üí aƒüƒ±rlƒ±k merkezi i√ßin gerekli
    M = cv2.moments(kontur)
    if M["m00"] != 0:
        cx = int(M["m10"] / M["m00"])  # x koordinatƒ±
        cy = int(M["m01"] / M["m00"])  # y koordinatƒ±
    else:
        cx, cy = 0, 0  # B√∂lme hatasƒ±na kar≈üƒ±

    print(f"Kontur {i+1}: Alan = {alan:.2f}, √áevre = {cevre:.2f}, Merkez = ({cx}, {cy})")
````

````
"""2. Kontur √ºst√ºnde Dikd√∂rtgen veya Daire √áizimi"""
kontur_kopya = cv2.cvtColor(normalize_gorsel_uint8, cv2.COLOR_GRAY2BGR)

for kontur in contours:
    # Dƒ±≈ü dikd√∂rtgen √ßiz
    x, y, w, h = cv2.boundingRect(kontur)
    cv2.rectangle(kontur_kopya, (x, y), (x + w, y + h), (0, 0, 255), 1)  # kƒ±rmƒ±zƒ±

    # En k√º√ß√ºk √ßevreleyen daire √ßiz
    (x_c, y_c), radius = cv2.minEnclosingCircle(kontur)
    center = (int(x_c), int(y_c))
    cv2.circle(kontur_kopya, center, int(radius), (255, 0, 0), 1)  # mavi

plt.figure(figsize=(10, 5))
plt.imshow(kontur_kopya)
plt.title("Konturlara Dikd√∂rtgen ve Daire √áizildi")
plt.axis("off")
plt.show()
````

######‚úÖ 1. Belirli Alanƒ±n √úst√ºndeki Nesneleri Se√ß,g√∂rselle≈ütir,ayrƒ± ayrƒ± kƒ±rp

***üéØ O zaman ≈üimdi g√∂r√ºnt√ºdeki nesneleri filtreleyerek analiz etmeye ge√ßiyoruz.***

Amacƒ±mƒ±z:

* K√º√ß√ºk nesneleri eleyerek sadece belirgin olanlarƒ± analiz etmek,

* Nesneleri saymak,

* ƒ∞steƒüe baƒülƒ± olarak her birini ayrƒ± ayrƒ± kƒ±rpmak.

````
#1. Belirli Alanƒ±n √ºst√ºndeki gorselleri sec

min_alan = 100  # Bu deƒüeri g√∂r√ºnt√ºye g√∂re ayarlayabilirsin

filtreli_konturlar = [cnt for cnt in contours if cv2.contourArea(cnt) > min_alan]

print(f"Toplam kontur: {len(contours)}")
print(f"Filtrelenmi≈ü (alan>{min_alan}) kontur sayƒ±sƒ±: {len(filtreli_konturlar)}")
````

````
#2. Secilen nesneleri gorsellestir.
gorsel_kopya = cv2.cvtColor(normalize_gorsel_uint8, cv2.COLOR_GRAY2BGR)
cv2.drawContours(gorsel_kopya, filtreli_konturlar, -1, (0, 255, 0), 1)

plt.imshow(gorsel_kopya)
plt.title(f"{len(filtreli_konturlar)} adet b√ºy√ºk nesne bulundu")
plt.axis("off")
plt.show()
````

````
#3. Her Nesneyi Ayrƒ± Ayrƒ± Kƒ±rp ve G√∂ster

for i, contur in enumerate(filtreli_konturlar):
    x, y, w, h = cv2.boundingRect(contur)
    kirpilmis = normalize_gorsel_uint8[y:y+h, x:x+w]

    plt.figure(figsize=(5,15))
    plt.imshow(kirpilmis, cmap='gray')
    plt.title(f"Nesne {i+1}")
    plt.axis("off")
    plt.show()
````

###### üéØ Hedef: Kƒ±rpƒ±lmƒ±≈ü nesneleri benzerliklerine g√∂re grupla
Y√∂ntem: G√∂rsel her nesnenin basitle≈ütirilmi≈ü √∂zelliklerini al, sonra benzer olanlarƒ± grupla. (K√ºmeleme = Clustering)

````
#HER NESNE ICIN OZELLIK CIKAR (ADIM YOGUNLUK)

ozellik_list=[]

for contur in filtreli_konturlar:
  x,y,w,h=cv2.boundingRect(contur)
  alan=cv2.contourArea(contur)
  #alan=w*h
  cevre=cv2.arcLength(contur,True)
  M=cv2.moments(contur)
  if M["m00"]!=0:
    cx=int(M["m10"]/M["m00"])
    cy=int(M["m01"]/M["m00"])
  ortalama_intensity=np.mean(normalize_gorsel_uint8[y:y+h,x:x+w])
  ozellik_list.append([alan,cevre,cx,cy,ortalama_intensity])

  print(ozellik_list)
````

````
#Adƒ±m 2: KMeans ile n_clusters=2(2 sƒ±nƒ±f ayƒ±rma)
from sklearn.cluster import KMeans

ozellikler = np.array(ozellik_list)
if len(filtreli_konturlar) >= 2:
    kmeans = KMeans(n_clusters=2).fit(ozellikler)
else:
    print("Yeterli sayƒ±da nesne yok, k√ºmeleme atlandƒ±.")

etiketler = kmeans.labels_

print(f"Toplam kontur sayƒ±sƒ±: {len(contours)}")
print(f"Filtreli kontur sayƒ±sƒ±: {len(filtreli_konturlar)}")
````

````
#Adƒ±m 4: Gruplara g√∂re renkli √ßizim
renkler = [(255, 0, 0), (0, 255, 0)]  # 2 sƒ±nƒ±f i√ßin

gorsel_kume = cv2.cvtColor(normalize_gorsel_uint8, cv2.COLOR_GRAY2BGR)

for idx, cnt in enumerate(filtreli_konturlar):
    x, y, w, h = cv2.boundingRect(cnt)
    renk = renkler[etiketler[idx]]
    cv2.rectangle(gorsel_kume, (x, y), (x+w, y+h), renk, 2)

plt.figure(figsize=(10, 5))
plt.imshow(gorsel_kume)
plt.title("K-means ile Sƒ±nƒ±flandƒ±rƒ±lmƒ±≈ü Nesneler")
plt.axis("off")
plt.show()
````

````
# Her Nesneyi Ayrƒ± G√∂rsel Dosyasƒ± Olarak Kaydetme (Gruplara G√∂re)
import os
from PIL import Image

# Klas√∂r olu≈ütur
os.makedirs("nesneler", exist_ok=True)

for idx, cnt in enumerate(filtreli_konturlar):
    x, y, w, h = cv2.boundingRect(cnt)
    roi = normalize_gorsel_uint8[y:y+h, x:x+w]

    grup_no = etiketler[idx]
    dosya_adi = f"nesneler/nesne_grup_{grup_no}_{idx}.png"

    # PIL ile kaydet (cv2.imwrite de olur ama PIL daha yaygƒ±n)
    im = Image.fromarray(roi)
    im.save(dosya_adi)

print("üéâ Nesneler dosyalara kaydedildi. Klas√∂r: 'nesneler'")
````

````
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
````
