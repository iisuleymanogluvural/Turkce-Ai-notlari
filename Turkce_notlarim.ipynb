````
#HESAP MAKINESI
def carpma(sayi1, sayi2):
    sonuc = 0.0
    for _ in range(int(sayi1)):
        sonuc += sayi2
    return sonuc

def karekok(x):
    tahmin = 0.0
    artim = 0.01
    while True:
        kare = carpma(tahmin, tahmin)
        if kare > x:
            break
        tahmin += artim
    return round(tahmin, 2)

def us_al(taban, us):
    tam_kisim = int(us)
    ondalik_kisim = us - tam_kisim

    # Tam kısmı hesapla
    sonuc = 1.0
    for _ in range(tam_kisim):
        sonuc = carpma(sonuc, taban)

    # Eğer ondalık kısım varsa (şimdilik sadece 0.5 destekliyoruz)
    if round(ondalik_kisim, 2) == 0.5:
        kok = karekok(taban)
        sonuc = carpma(sonuc, kok)

    return round(sonuc, 2)

# Örnekler
print(us_al(4, 1.5))  # 4^1.5 = 4 * sqrt(4) = 4 * 2 = 8
print(us_al(9, 0.5))  # sqrt(9) = 3
print(us_al(2, 3.5))  # 2^3 * sqrt(2) = 8 * ~1.41 = ~11.3

#LIST COMPREHENSION
print([i*i for i in range(0,10,2)])
print([i / 10 if i * 2 > 10 else 0 for i in range(0, 10, 2)])
````

###🧠 Yeni Konu: Basit Görüntü Filtresi (Bulanıklaştırma / Blur)
NumPy kullanarak 2D görsel üzerine ortalama filtresi uygulayacağız. Bu, çevresindeki piksellerin ortalamasını alarak görseli yumuşatır (bulanıklaştırır).



````
# once fonksiyonu tanımla:
def blur(image):
  image = image.astype(np.float32)
  blurred_image=np.copy(image)
  height,width=image.shape
  for i in range(1,height-1):
    for j in range(1,width-1):
      blurred_image[i,j]=np.mean(image[i-1:i+2,j-1:j+2])
  return blurred_image

#filtreyi uygula ve gorsellestir.
blur_gorsel=blur(normalize_gorsel)
plt.imshow(blur_gorsel,cmap="gray")
plt.title("Blur Görüntü")
plt.colorbar()
plt.show()
````

####🔄 İstersen bunları da yapabilirsin (bonus)
📌 Kenar tespiti için np.abs(image[i+1, j] - image[i, j]) gibi farklı filtreler dene.

📈 Filtre öncesi ve sonrası ortalama, std gibi metrikleri karşılaştır.

kenar tespiti (edge detection) mantığını anlatacağım ve np.abs(image[i+1, j] - image[i, j]) satırını doğru şekilde bir filtre fonksiyonu içine yerleştireceğiz.

🎯 Amaç:
Pikselin kendisi ile altındaki piksel arasındaki farkı alarak kenarları ortaya çıkarmak.

######✅ 1. Kenar Tespiti için Basit Fark Filtresi
Bu filtre, pikseller arasındaki ani değişimleri tespit eder — bu da genellikle bir kenar anlamına gelir.

````
def kose_tanimlama(image):
    image = image.astype(np.float32)
    kenar_tanimlama = np.zeros_like(image)
    height, width = image.shape

    for i in range(1, height - 1):
        for j in range(1, width - 1):
            dikey_fark = np.abs(image[i+1, j] - image[i, j])
            yatay_fark = np.abs(image[i, j+1] - image[i, j])
            kenar_tanimlama[i, j] = max(dikey_fark, yatay_fark)

    return kenar_tanimlama

# Fonksiyonu uygula
kenarlar = kose_tanimlama(normalize_gorsel)

# Görselleştir
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1) #1satır : 3 sütun 1.grafik
plt.imshow(normalize_gorsel, cmap='gray')
plt.title("Orijinal Görüntü")

plt.subplot(1, 3, 2) #1 satır :3 sütun 2.grafik
plt.imshow(blur_gorsel, cmap='gray')
plt.title("Bulanık Görüntü")

plt.subplot(1, 3, 3) #1 satır :3 sütun 3. grafik
plt.imshow(kenarlar, cmap='gray')
plt.title("Kenar Tespit")

plt.tight_layout() #üst üste binme düzensizlik varsa düzeltir
plt.show()

# İstatistik
print("Bulanık Görüntü Ortalaması:", blur_gorsel.mean())
print("Bulanık Görüntü Std Sapması:", blur_gorsel.std())

"""Eğer kod çalışmadıysa, muhtemelen şu sorunlardan biri oluşmuştur:

Görselin (image) tipi int olduğu için, ortalamaları alırken tam sayı bölme hatası olabilir.

blurred[i, j] = np.mean(...) ifadesi float döndürür, ama blurred hala int dizisi olduğu için tip uyuşmazlığı olabilir.

Kod biraz yavaş çalışabilir çünkü iç içe döngü var."""
````

######🧠 NEDİR: Eşikleme (Thresholding)?
Eşikleme, piksel değerlerinin belirli bir sınırın altında veya üstünde olup olmadığına göre işlem yapılmasıdır.

Örneğin:

Eğer kenar kuvveti 0.1’den büyükse: "bu bir kenardır"

Aksi halde: "kenar değil"



````
def kenar_esikleme(edge_image, threshold=0.1):
    binary_koseler = np.zeros_like(edge_image)
    binary_koseler[edge_image > threshold] = 1.0
    return binary_koseler

kenar_gri = kose_tanimlama(normalize_gorsel)  # Önce kenarları bul
kenar_belirgin = kenar_esikleme(kenar_gri, threshold=0.08)  # Eşik uygula

plt.imshow(kenar_belirgin, cmap='gray')
plt.title("Eşiklenmiş Kenarlar")
plt.show()
````

````
# BONUS : Kontur ile KENAR ÇİZİMİ

plt.contour(kenar_gri, levels=[0.1], colors='red')
plt.title("Kontur ile Kenar Çizimi")
plt.show()
````

````
def kontrast_arttir(gorsel):
    return (gorsel - gorsel.min()) / (gorsel.max() - gorsel.min())
    return kontrast_arttir

plt.imshow(kontrast_arttir(normalize_gorsel),cmap="gray")
plt.title("Kontrastlı Görüntü")
plt.colorbar()
plt.show()
````

######🔍 Sobel Filtresi ile Gelişmiş Kenar Tespiti
📌 Nedir Bu?
Sobel filtresi, hem yatay hem de dikey kenarları aynı anda daha net tespit etmek için kullanılır.
Basit fark alma yöntemlerinden daha güçlüdür ve görüntüdeki ani değişimleri daha iyi algılar.

🧠 Temel Mantık:
Sobel, iki yönlü filtre uygular:


````
import matplotlib.pyplot as plt
from scipy.ndimage import convolve

# Sobel filtre çekirdekleri
sobel_x = np.array([[ -1, 0, 1],
                    [ -2, 0, 2],
                    [ -1, 0, 1]])

sobel_y = np.array([[ -1, -2, -1],
                    [  0,  0,  0],
                    [  1,  2,  1]])
#Bu matrisler, görüntüye konvolüsyon işlemi (kaydırılarak çarpma) uygulanarak kenar bilgisi çıkarılır.

def sobel_kenar(image):
    image = image.astype(np.float32)
    gx = convolve(image, sobel_x)
    gy = convolve(image, sobel_y)
    sobel_magnitude = np.sqrt(gx**2 + gy**2)
    return sobel_magnitude / sobel_magnitude.max()  # normalize et

sobel_sonuc = sobel_kenar(normalize_gorsel)

plt.imshow(sobel_sonuc, cmap='gray')
plt.title("Sobel Kenar Tespiti")
plt.axis("off")
plt.show()
````

🧠 Bilmen Gerekenler
* Kavram	Açıklama
* convolve	== Görüntüye filtreyi uygular (kaydırmalı çarpma)
* sobel_x, sobel_y	==Görüntüde yönlü değişimi bulur
* sqrt(gx² + gy²)	==Gerçek kenar büyüklüğü
* normalize ==	Görseli 0-1 arası parlaklığa getirir

🧠 Sorunun Sebebi:
* Senin görselin büyük ihtimalle gürültü (noise) içeriyor — yani piksel değerleri rastgele dağılmış. Bu tür görüntülerde Sobel gibi kenar tespit yöntemleri işe yaramaz çünkü:

** Gerçek kenar yerine gürültüye tepki verir.

** Kenar belirginliği az olur.

* ✅ Çözüm: Görseli Bulanıklaştır (Smoothing)
Sobel'den önce, görüntüyü yumuşatarak (bulanıklaştırarak) gürültüyü bastırmalıyız.

######🔍 Canny Kenar Tespiti Nedir?
* Canny algoritması, görüntüdeki en doğru ve keskin kenarları bulmak için geliştirilmiş bir yöntemdir. Sobel'den daha akıllıdır çünkü:

````
#1 QAUSS FILTRESI UYGULAMA
from scipy.ndimage import gaussian_filter

blurred = gaussian_filter(normalize_gorsel, sigma=1.0)

plt.imshow(blurred, cmap='gray')
plt.title(" GAUSS FILTRESI ILE Bulanıklaştırılmış Görüntü ")
plt.colorbar()
plt.show()
````

````
#2 SOBEL ILE KENAR GUCU HESABI
from scipy.ndimage import convolve

# Sobel filtre çekirdekleri
sobel_x = np.array([[ -1, 0, 1],
                    [ -2, 0, 2],
                    [ -1, 0, 1]])

sobel_y = np.array([[ -1, -2, -1],
                    [  0,  0,  0],
                    [  1,  2,  1]])
#Bu matrisler, görüntüye konvolüsyon işlemi (kaydırılarak çarpma) uygulanarak kenar bilgisi çıkarılır.

def sobel_kenar(image):
    image = image.astype(np.float32)
    gx = convolve(image, sobel_x)
    gy = convolve(image, sobel_y)
    sobel_magnitude = np.sqrt(gx**2 + gy**2)
    return sobel_magnitude / sobel_magnitude.max()  # normalize et

sobel_sonuc = sobel_kenar(blurred)

plt.imshow(sobel_sonuc, cmap='gray')
plt.title("Sobel Kenar Tespiti")
plt.axis("off")
plt.show()
````

````
#3 THRESHOLDING ile zayif-kuvvetli kenar ayrilir.
def kenar_esikleme(edge_image, threshold=0.1):
    binary_koseler = np.zeros_like(edge_image)
    binary_koseler[edge_image > threshold] = 1.0
    return binary_koseler

kenar_gri = kose_tanimlama(sobel_sonuc)  # Önce kenarları bul
kenar_belirgin = kenar_esikleme(kenar_gri, threshold=0.08)  # Eşik uygula

plt.imshow(kenar_belirgin, cmap='gray')
plt.title("Eşiklenmiş Kenarlar")
plt.show()
````

````
#Non-maximum suppression (inceleştirme) yapılır
"""NMS, kenarları inceltmek için kullanılır.
Amaç: Kenar boyunca sadece en güçlü pikseli (maksimum gradyan) bırakmak, diğerlerini sıfırlamak.
Sobel veya diğer gradyan filtreleri kenar boyunca geniş alanlar verir — kalın çizgiler.
Ama biz yalnızca tam kenar hattını istiyoruz.
Bu yüzden, kenar boyunca sadece en yüksek değeri koruruz, diğerlerini sıfırlarız."""

def non_maximum_suppression(magnitude, gx, gy):
    height, width = magnitude.shape
    suppressed = np.zeros_like(magnitude)

    angle = np.arctan2(gy, gx) * 180.0 / np.pi
    angle[angle < 0] += 180  # Açıyı 0-180 aralığına getir

    for i in range(1, height - 1):
        for j in range(1, width - 1):
            q = 255
            r = 255

            # 0 derece
            if (0 <= angle[i, j] < 22.5) or (157.5 <= angle[i, j] <= 180):
                q = magnitude[i, j + 1]
                r = magnitude[i, j - 1]
            # 45 derece
            elif (22.5 <= angle[i, j] < 67.5):
                q = magnitude[i + 1, j - 1]
                r = magnitude[i - 1, j + 1]
            # 90 derece
            elif (67.5 <= angle[i, j] < 112.5):
                q = magnitude[i + 1, j]
                r = magnitude[i - 1, j]
            # 135 derece
            elif (112.5 <= angle[i, j] < 157.5):
                q = magnitude[i - 1, j - 1]
                r = magnitude[i + 1, j + 1]

            if (magnitude[i, j] >= q) and (magnitude[i, j] >= r):
                suppressed[i, j] = magnitude[i, j]
            else:
                suppressed[i, j] = 0

    return suppressed

# Sobel yönlerini de alacak şekilde fonksiyonu güncelle
def sobel_kenar_yonlu(image):
    image = image.astype(np.float32)
    gx = convolve(image, sobel_x)
    gy = convolve(image, sobel_y)
    magnitude = np.sqrt(gx**2 + gy**2)
    return magnitude / magnitude.max(), gx, gy

# Kenar gücü ve yönlerini al
sobel_magnitude, gx, gy = sobel_kenar_yonlu(blurred)

# NMS uygula
nms_sonuc = non_maximum_suppression(sobel_magnitude, gx, gy)

# Görüntüle
plt.imshow(nms_sonuc, cmap='gray')
plt.title("Non-Maximum Suppression Sonucu")
plt.axis("off")
plt.show()
````

````
#Edge tracking by hysteresis (kenarları birbirine bağlama) uygulanır
"""Nedir Bu "Hysteresis"?
Bu adımda, elimizdeki kenarları güçlü ve zayıf olarak sınıflandırıyoruz:

Güçlü kenar: Yüksek değere sahip — bunlar kesin kenardır.

Zayıf kenar: Daha düşük değere sahip — ama eğer güçlü kenara bağlıysa, onu da tut."""

def hysteresis_thresholding(img, low_thresh=0.05, high_thresh=0.15):
    strong = 1.0
    weak = 0.5

    result = np.zeros_like(img)

    strong_i, strong_j = np.where(img >= high_thresh)
    weak_i, weak_j = np.where((img >= low_thresh) & (img < high_thresh))

    result[strong_i, strong_j] = strong
    result[weak_i, weak_j] = weak

    # Komşulukla zayıf kenarları güçlüye bağlıysa tut
    for i in range(1, img.shape[0]-1):
        for j in range(1, img.shape[1]-1):
            if result[i, j] == weak:
                # 8 komşusuna bak
                if ((result[i+1, j-1] == strong) or (result[i+1, j] == strong) or (result[i+1, j+1] == strong)
                    or (result[i, j-1] == strong) or (result[i, j+1] == strong)
                    or (result[i-1, j-1] == strong) or (result[i-1, j] == strong) or (result[i-1, j+1] == strong)):
                    result[i, j] = strong
                else:
                    result[i, j] = 0
    return result

hysteresis_sonuc = hysteresis_thresholding(nms_sonuc, low_thresh=0.05, high_thresh=0.15)

plt.imshow(hysteresis_sonuc, cmap='gray')
plt.title("Hysteresis Thresholding Sonucu (Canny'nin Finali- ama el ile kodlanmış)")
plt.axis("off")
plt.show()
````

````
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Kontur bul
contours, _ = cv2.findContours(canny_kenarlar, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Konturları orijinal görüntü üzerine çizmek için bir kopya
kontur_gorsel = cv2.cvtColor(normalize_gorsel_uint8, cv2.COLOR_GRAY2BGR)
cv2.drawContours(kontur_gorsel, contours, -1, (0, 255, 0), 1)

for i, cnt in enumerate(contours[:10]):  # İlk 10 kontur
    alan = cv2.contourArea(cnt)
    cevre = cv2.arcLength(cnt, True)
    print(f"Kontur {i+1}: Alan={alan:.2f}, Çevre={cevre:.2f}")

# Görselleştir
plt.figure(figsize=(10,5))
plt.imshow(kontur_gorsel)
plt.title(f"{len(contours)} adet kontur bulundu")
plt.axis("off")
plt.show()
````

######✅ 1. Kontur Alanı, Çevresi ve Merkezi Hesaplama, Kontur üstünde Dikdörtgen veya Daire Çizimi
***🧠 Ne Öğrendik?***
* **cv2.contourArea:** Alan hesaplar.

* **cv2.arcLength:** Çevre hesaplar.

* **cv2.boundingRect:** En küçük dikdörtgen.

* **cv2.minEnclosingCircle:** En küçük çember.

* **cv2.moments:** Nesnenin merkezini verir.

````
"""1. Kontur Alanı, Çevresi ve Merkezi Hesaplama"""

for i, kontur in enumerate(contours):
    alan = cv2.contourArea(kontur)
    cevre = cv2.arcLength(kontur, True)

    # M (moment) hesapla → ağırlık merkezi için gerekli
    M = cv2.moments(kontur)
    if M["m00"] != 0:
        cx = int(M["m10"] / M["m00"])  # x koordinatı
        cy = int(M["m01"] / M["m00"])  # y koordinatı
    else:
        cx, cy = 0, 0  # Bölme hatasına karşı

    print(f"Kontur {i+1}: Alan = {alan:.2f}, Çevre = {cevre:.2f}, Merkez = ({cx}, {cy})")
````

````
"""2. Kontur üstünde Dikdörtgen veya Daire Çizimi"""
kontur_kopya = cv2.cvtColor(normalize_gorsel_uint8, cv2.COLOR_GRAY2BGR)

for kontur in contours:
    # Dış dikdörtgen çiz
    x, y, w, h = cv2.boundingRect(kontur)
    cv2.rectangle(kontur_kopya, (x, y), (x + w, y + h), (0, 0, 255), 1)  # kırmızı

    # En küçük çevreleyen daire çiz
    (x_c, y_c), radius = cv2.minEnclosingCircle(kontur)
    center = (int(x_c), int(y_c))
    cv2.circle(kontur_kopya, center, int(radius), (255, 0, 0), 1)  # mavi

plt.figure(figsize=(10, 5))
plt.imshow(kontur_kopya)
plt.title("Konturlara Dikdörtgen ve Daire Çizildi")
plt.axis("off")
plt.show()
````

######✅ 1. Belirli Alanın Üstündeki Nesneleri Seç,görselleştir,ayrı ayrı kırp

***🎯 O zaman şimdi görüntüdeki nesneleri filtreleyerek analiz etmeye geçiyoruz.***

Amacımız:

* Küçük nesneleri eleyerek sadece belirgin olanları analiz etmek,

* Nesneleri saymak,

* İsteğe bağlı olarak her birini ayrı ayrı kırpmak.

````
#1. Belirli Alanın üstündeki gorselleri sec

min_alan = 100  # Bu değeri görüntüye göre ayarlayabilirsin

filtreli_konturlar = [cnt for cnt in contours if cv2.contourArea(cnt) > min_alan]

print(f"Toplam kontur: {len(contours)}")
print(f"Filtrelenmiş (alan>{min_alan}) kontur sayısı: {len(filtreli_konturlar)}")
````

````
#2. Secilen nesneleri gorsellestir.
gorsel_kopya = cv2.cvtColor(normalize_gorsel_uint8, cv2.COLOR_GRAY2BGR)
cv2.drawContours(gorsel_kopya, filtreli_konturlar, -1, (0, 255, 0), 1)

plt.imshow(gorsel_kopya)
plt.title(f"{len(filtreli_konturlar)} adet büyük nesne bulundu")
plt.axis("off")
plt.show()
````

````
#3. Her Nesneyi Ayrı Ayrı Kırp ve Göster

for i, contur in enumerate(filtreli_konturlar):
    x, y, w, h = cv2.boundingRect(contur)
    kirpilmis = normalize_gorsel_uint8[y:y+h, x:x+w]

    plt.figure(figsize=(5,15))
    plt.imshow(kirpilmis, cmap='gray')
    plt.title(f"Nesne {i+1}")
    plt.axis("off")
    plt.show()
````

###### 🎯 Hedef: Kırpılmış nesneleri benzerliklerine göre grupla
Yöntem: Görsel her nesnenin basitleştirilmiş özelliklerini al, sonra benzer olanları grupla. (Kümeleme = Clustering)

````
#HER NESNE ICIN OZELLIK CIKAR (ADIM YOGUNLUK)

ozellik_list=[]

for contur in filtreli_konturlar:
  x,y,w,h=cv2.boundingRect(contur)
  alan=cv2.contourArea(contur)
  #alan=w*h
  cevre=cv2.arcLength(contur,True)
  M=cv2.moments(contur)
  if M["m00"]!=0:
    cx=int(M["m10"]/M["m00"])
    cy=int(M["m01"]/M["m00"])
  ortalama_intensity=np.mean(normalize_gorsel_uint8[y:y+h,x:x+w])
  ozellik_list.append([alan,cevre,cx,cy,ortalama_intensity])

  print(ozellik_list)
````

````
#Adım 2: KMeans ile n_clusters=2(2 sınıf ayırma)
from sklearn.cluster import KMeans

ozellikler = np.array(ozellik_list)
if len(filtreli_konturlar) >= 2:
    kmeans = KMeans(n_clusters=2).fit(ozellikler)
else:
    print("Yeterli sayıda nesne yok, kümeleme atlandı.")

etiketler = kmeans.labels_

print(f"Toplam kontur sayısı: {len(contours)}")
print(f"Filtreli kontur sayısı: {len(filtreli_konturlar)}")
````

````
#Adım 4: Gruplara göre renkli çizim
renkler = [(255, 0, 0), (0, 255, 0)]  # 2 sınıf için

gorsel_kume = cv2.cvtColor(normalize_gorsel_uint8, cv2.COLOR_GRAY2BGR)

for idx, cnt in enumerate(filtreli_konturlar):
    x, y, w, h = cv2.boundingRect(cnt)
    renk = renkler[etiketler[idx]]
    cv2.rectangle(gorsel_kume, (x, y), (x+w, y+h), renk, 2)

plt.figure(figsize=(10, 5))
plt.imshow(gorsel_kume)
plt.title("K-means ile Sınıflandırılmış Nesneler")
plt.axis("off")
plt.show()
````

````
# Her Nesneyi Ayrı Görsel Dosyası Olarak Kaydetme (Gruplara Göre)
import os
from PIL import Image

# Klasör oluştur
os.makedirs("nesneler", exist_ok=True)

for idx, cnt in enumerate(filtreli_konturlar):
    x, y, w, h = cv2.boundingRect(cnt)
    roi = normalize_gorsel_uint8[y:y+h, x:x+w]

    grup_no = etiketler[idx]
    dosya_adi = f"nesneler/nesne_grup_{grup_no}_{idx}.png"

    # PIL ile kaydet (cv2.imwrite de olur ama PIL daha yaygın)
    im = Image.fromarray(roi)
    im.save(dosya_adi)

print("🎉 Nesneler dosyalara kaydedildi. Klasör: 'nesneler'")
````

## Veri bilimi
</div>

Colab ile popüler Python kitaplıklarının tüm avantajlarından yararlanarak veri analiz edip görselleştirebilirsiniz. Aşağıdaki kod hücresi rastgele veri oluşturmak için <strong>numpy</strong>'yi, bu veriyi görselleştirmek için de <strong>matplotlib</strong>'i kullanır. Kodu düzenlemek için hücreyi tıklamanız yeterlidir. Sonrasında düzenlemeye başlayabilirsiniz.

Kendi verilerinizi Google Drive hesabınızdan &#40;e-tablolar dahil&#41;, GitHub'dan ve diğer pek çok kaynaktan Colab not defterlerine aktarabilirsiniz. Veri içe aktarma ve Colab'in veri bilimi için nasıl kullanılabileceği hakkında daha fazla bilgi edinmek için <a href="#working-with-data">Verilerle Çalışma</a> bölümünün altındaki bağlantılara bakabilirsiniz.

````
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
````
